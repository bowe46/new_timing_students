```python
def b(e, kt):
    return ( 2 *np.pi * (e * u.keV)**3 / (cons.h * cons.c)**2 * 1/ ( np.exp (e/kt ) -1 ) ).cgs.value
```

```python
e=10**np.linspace(-1.,2., 100)
kt =2.
spec=b(e,kt)
```

```python
plt.figure()
plt.loglog(e,(spec))
plt.xlabel('Energy [keV]')
plt.ylabel('B(E,kT=%.1f keV) [g/s^2]'%kt)
```

```python
import scipy.integrate as integrate

low=0.1
high=10.0
result = integrate.quad(lambda x: b(x,kt), low,high) 
lumin= result[0] * (u.keV/cons.h).cgs.value * 4* np.pi * ((10 * u.km).cgs.value)**2

print('L(%.1f-%.1f) = %.2g erg/s'%(low,high, lumin))
```
